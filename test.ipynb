```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
kanas = 'あいうえおかきくけこさしすせそたちつてとなにぬねのはひふへほまみむめもやゆよらりるれろわをん'
goju = [
    ['あ', 'い', 'う', 'え', 'お'],
    ['か', 'き', 'く', 'け', 'こ'],
    ['さ', 'し', 'す', 'せ', 'そ'],
    ['た', 'ち', 'つ', 'て', 'と'],
    ['な', 'に', 'ぬ', 'ね', 'の'],
    ['は', 'ひ', 'ふ', 'へ', 'ほ'],
    ['ま', 'み', 'む', 'め', 'も'],
    ['や', None, 'ゆ', None, 'よ'],
    ['ら', 'り', 'る', 'れ', 'ろ'],
    ['わ', None, 'を', None, 'ん']]
```

```python
def img_to_arr(img):
    img = img.convert('L')
    img = img.resize((32, 32))
    return (255 - np.array(img)) / 255

for img_id in range(5):
    fig = plt.figure(figsize=(10,20))
    for dan_id, dan in enumerate(goju):
        for gyo_id, char in enumerate(dan):
            if char is not None:
                fig.add_subplot(10, 5, dan_id * 5 + gyo_id + 1)
                img = img_to_arr(Image.open('image{}/{}.png'.format(img_id + 1, char)))
                plt.imshow(img, cmap='binary')
    plt.show()
```
